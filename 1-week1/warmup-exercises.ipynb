2. Read the `titanic.csv` into Pandas.
   a. Remove rows with missing values.
   b. Using a single train / test split, create a decision tree classifier to predict "survived"

```python
#Question 2 Answer
data_q2=pd.read_csv('data/titanic.csv')
# a.
data_q2=data_q2.dropna(how='any')
# drop name column for modeling
data_q2=data_q2.drop(['name'],axis=1)
cols=data_q2.select_dtypes(include='object').columns
dummies=pd.get_dummies(data_q2[cols])
data_q2.drop(cols, axis=1, inplace=True)
data_q2=pd.concat([data_q2,dummies],axis=1)
data_q2
# train test split
from sklearn.model_selection import train_test_split
Y=data_q2.survived
X=data_q2.iloc[:,0:10]
x_train, x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=2)
## decision tree classifier
from sklearn.tree import DecisionTreeClassifier
tc = DecisionTreeClassifier() 
tc.fit(x_train,y_train)
y_test_pred=pd.Series(tc.predict(x_test))
y_test.index=y_test_pred.index
result_table=pd.concat([y_test_pred,y_test],axis=1)
result_table
```
